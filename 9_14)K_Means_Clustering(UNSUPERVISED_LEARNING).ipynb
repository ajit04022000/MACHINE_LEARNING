```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv')
```

```python
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
plt.xlabel('Age')
plt.ylabel('income')
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
# in data set we can see that the name is not important that much for making cluster
# and KMeans helps to do "fit_predict" in one time
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['Cluster'] = y_predicted
df.head()
```

```python
# these are the centroide of our 3 clusters
km.cluster_centers_
```

```python
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
df3=df[df.Cluster==2]
```

```python
plt.scatter(df1.Age,df1['Income($)'],color = 'green')
plt.scatter(df2.Age,df2['Income($)'],color = 'red')
plt.scatter(df3.Age,df3['Income($)'],color = 'black')
plt.xlabel('Age')
plt.ylabel('income')
plt.legend()
```

so for that we are goinging to scalling all the data by the help of "MinMaxScallar".

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(df[['Income($)']])
df['Income($)']= scalar.transform(df[['Income($)']])
scalar.fit(df[['Age']])
df['Age']= scalar.transform(df[['Age']])
df.head()
```

by the help of MinMaxScalar the "Income($)" values are transform into 0 to 1 range.

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
# in data set we can see that the name is not important that much for making cluster
# and KMeans helps to do "fit_predict" in one time
y_predicted = km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['Cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
df3=df[df.Cluster==2]
```

```python
plt.scatter(df1.Age,df1['Income($)'],color = 'green')
plt.scatter(df2.Age,df2['Income($)'],color = 'red')
plt.scatter(df3.Age,df3['Income($)'],color = 'black')
plt.xlabel('Age')
plt.ylabel('income')
plt.legend()
```

```python
# ploting all the value with CENTROID
plt.scatter(df1.Age,df1['Income($)'],color = 'green')
plt.scatter(df2.Age,df2['Income($)'],color = 'red')
plt.scatter(df3.Age,df3['Income($)'],color = 'black')
plt.xlabel('Age')
plt.ylabel('income')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

##ELBOW PLOT

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```

1. Use iris flower dataset from sklearn library and try to form clusters of flowers using petal width and length features. Drop other two features for simplicity.

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['flower'] = iris.target
df.head()
```

```python
df.drop(['sepal length (cm)', 'sepal width (cm)', 'flower'],axis='columns',inplace=True)
```

```python
df.head(3)
```

```python
km = KMeans(n_clusters=3)
yp = km.fit_predict(df)
yp
```

```python
df['cluster'] = yp
df.head(2)
```

```python
df.cluster.unique()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='green')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'],color='yellow')
```

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
