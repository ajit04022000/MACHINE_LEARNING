```python
import pandas as pd
import numpy as numpy
from sklearn import linear_model
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# List of names
area = [2600,3000,3200,3600,4000]
price = [550000,565000,610000,680000,725000]  
      
# Get the list of tuples by zipping the two lists together   
list_of_tuples = list(zip(area, price))  
      
# Converting the list of tuples into pandas Dataframe 
df = pd.DataFrame(list_of_tuples, columns = ['area', 'price'])
     
# Print the dataframe 
df
```

```python
#making a scatter plot to visualize the data
plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel('area')
plt.ylabel('price')
```

```python
# making a new data set and droping thhe price column
# because the fit _method alwas accepect a 2d array
new_df = df.drop('price',axis=1)
new_df
```

```python
model= linear_model.LinearRegression()
model.fit(new_df,df.price)
```

```python
# finding the 3300 sqrt house price
model.predict([[3300]])
```

```python
# finding the 5000 sqrt house price
model.predict([[5000]])
```

```python
# finding the slope of the model (y= mx + c)  hear the slope is "m"
model.coef_
```

```python
# finding the intercept of the model "c"
model.intercept_
```

```python
# hear we can see that we have the intercept and cofficent value prasent
# so now if we putt all the value to the formula(y= mx + c) we find the beat line
y = 135.78767123*5000+180616.43835616432
print(y)
print('so hear we can se that our predicting well and throw the same value')
```

```python
# this is our dataset contain al the area which we want to predict 
area = [1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000]
area_df = pd.DataFrame(area,columns=['area'])
area_df
```

```python
#we are predicting the price for all the values of area are present
price = model.predict(area_df)
price
```
