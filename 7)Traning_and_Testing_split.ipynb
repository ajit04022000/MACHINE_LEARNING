```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/6_train_test_split/carprices.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(data= df,x='Mileage',y= 'Sell Price($)')
```

```python
plt.scatter(data= df,x='Age(yrs)',y= 'Sell Price($)')
```

it's lienear in nature so we easily fit the linear regression on this data set

```python
X = df[['Mileage','Age(yrs)']]
y = df['Sell Price($)']
```

```python
X
```

```python
y
```

```python
# using sklearn 's train test split to split the model into traning and testing groups
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
print("the shape of x_train",len(X_train))
print("the shape of X_test",len(X_test))
print("the shape of y_train",len(y_train))
print("the shape of y_test",len(y_test))
```

```python
# using linear regrassion to train my data set
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
X_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
